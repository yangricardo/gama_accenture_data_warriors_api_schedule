## Função de exibição de gráfico de linha por continente (5 maiores)

```
def plot_line(cont):
    df_daily_dash=df_daily.toPandas()
    df_summary_dash=df_summary.toPandas()
    summary=df_summary_dash[df_summary_dash["continentname"]==cont]
    summary=summary.sort_values(by=["totalconfirmed"],ascending=False)
    summary=summary.head(5).copy()
    lista_paises=list(summary["countrycode"])
    base=df_daily_dash[(df_daily_dash["continentname"]==cont) & (df_daily_dash["countrycode"].isin(lista_paises))]
    #sns.lineplot(x="datereg", y="newdeaths", data=base[base['countrycode']!='FR'],hue="countryname")
    return px.line(x="datereg", y="newdeaths", color="countryname",data_frame=base, title=f"Daily deaths {cont}"
)
```

## 3 Proporção do número de casos por continente

```
#from plotly.offline import plot
#import plotly.graph_objs as go
#import plotly.express as px
pandas_df = df_daily.toPandas()
dic=pandas_df.to_dict("records")
title="Proporção de número de casos por continente"
px.treemap(pandas_df, path=["continentname","countryname"], values='totalconfirmed',
                  color_continuous_scale='RdBu',
                  range_color=[0,1],
                  title=title)

```

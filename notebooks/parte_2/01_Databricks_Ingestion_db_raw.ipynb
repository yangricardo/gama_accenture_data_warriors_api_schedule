# **Ingestão do Banco PostgreSQL para Stage Raw (.json)**

####1. Definindo função para carregamento dos dataframes

```
def loadSparkDataframeFromSQLTable(
    dbtable,
    driver="org.postgresql.Driver",
    url="jdbc:postgresql://data-warriors-postgresql.postgres.database.azure.com:5432/postgres",
    user="DataWarriorsAdmin@data-warriors-postgresql",
    password="DataWarriorsPassword!",
    sslmode="require"):
  return spark.read \
    .format("jdbc") \
    .option("driver", driver)\
    .option("url", url) \
    .option("dbtable", dbtable) \
    .option("user", user) \
    .option("password", password) \
    .option("sslmode", sslmode) \
    .load()
```

####2.Criando os dataframes correnpondentes as tabelas do Banco Postgres

```
country_df = loadSparkDataframeFromSQLTable('country')
daily_df = loadSparkDataframeFromSQLTable('daily')
summary_df = loadSparkDataframeFromSQLTable('summary')
continent_df = loadSparkDataframeFromSQLTable('continent')
country_continent_df = loadSparkDataframeFromSQLTable('country_continent')
```

```
country_iso_df=spark.read.load('dbfs:/FileStore/tables/country_code_iso-2.csv',format='csv',header='true',sep=';')
```

####3. Salvando os dataframes em .jon no  dir [dbfs:/FileStore/covid_data_lake/raw](/dbfs/FileStore/covid_data_lake/raw)

```
def save_raw_data(df, file_name):
  df.write\
    .mode('overwrite')\
    .option('header', 'true')\
    .json(f'dbfs:/FileStore/covid_data_lake/raw/db_raw_{file_name}_json')
```

```
save_raw_data(country_df, 'country')
save_raw_data(daily_df, 'daily')
save_raw_data(summary_df, 'summary')
save_raw_data(continent_df, 'continent')
save_raw_data(country_continent_df, 'country_continent')
save_raw_data(country_iso_df,'country_iso')
```
